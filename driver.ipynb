### Item Data Request
Filtering on items only used in

gamemode=CLASSIC (Classic Summoner's Rift)  
mapId=11 (Current Summoner's Rift)

based on information from  
https://static.developer.riotgames.com/docs/lol/gameModes.json  
https://static.developer.riotgames.com/docs/lol/maps.json

```python
item_data_request = requests.get("http://ddragon.leagueoflegends.com/cdn/" + latest_version + "/data/en_US/item.json")
item_data = item_data_request.json()['data']
```

```python
item_df = pd.DataFrame(item_data)
item_df = item_df.transpose()

# Explode dictionaries inside columns
explode_columns = ['stats','maps','gold']
for col in explode_columns:
    item_df = pd.concat([item_df.drop([col], axis=1), pd.json_normalize(item_df[col]).set_index(item_df.index)], axis=1)

# Filter on items used in Summoner's Rift maps and purchasable
item_df = item_df[item_df['11'] == True]
item_df = item_df[item_df['purchasable'] == True]

# Drop non-relevant data columns
item_df = item_df.drop(['description', 'colloq', 'plaintext', 'image'], axis=1)
item_df = item_df.drop(['11', '12', '21','22','purchasable'], axis=1)
item_df = item_df.drop(['effect'], axis=1) # Dropping effect for reduced complexity

for col in item_df.columns[4:]:
    if list(item_df[col].unique()) == [np.nan]:
        item_df = item_df.drop([col], axis=1)

# Fill NaN values in boolean types to False
item_df['consumed'].fillna(False, inplace=True)
item_df['hideFromAll'].fillna(False, inplace=True)
item_df['consumeOnFull'].fillna(False, inplace=True)

# Fill NaN values in stat modifiers to 0
for col in ['FlatMovementSpeedMod',
       'FlatHPPoolMod', 'FlatCritChanceMod', 'FlatMagicDamageMod',
       'FlatMPPoolMod', 'FlatArmorMod', 'FlatSpellBlockMod',
       'FlatPhysicalDamageMod', 'PercentAttackSpeedMod', 'PercentLifeStealMod',
       'FlatHPRegenMod', 'PercentMovementSpeedMod']:
    item_df[col].fillna(0, inplace=True)


item_df
```
